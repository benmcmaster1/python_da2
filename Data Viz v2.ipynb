```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your graph. Below size is DPI x 5 wide (500px) and DPI x 3 high (300px)
plt.figure(figsize=(5,3), dpi=100)

#plt.plot(x,y, label='2x', c='#ff00ff', linewidth=3, marker='o', ms=12, linestyle='--', markeredgecolor='black')

## Line number one.
plt.plot(x,y, 'g.--', label='2x')

## Line number two. How to show a 'projection' with dotted lines starting half way through.
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title("Our first graph", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("The X Axis", fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel("The Y Axis")
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

plt.figure(figsize=(6,4), dpi=110)

bars = plt.bar(labels,values)

patterns = ['/', 'O', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
#bars[0].set_hatch('/')
#bars[1].set_hatch('|')
#bars[-1].set_hatch('O')

plt.show()
```
